## Handling Mssing values

```python
data.isnull().sum()
```

```python
data["Shipment Mode"] = data["Shipment Mode"].fillna("Missing")
```

```python
data["Dosage"] = data["Dosage"].fillna("Missing")
```

```python
data["Line Item Insurance (USD)"] =  data["Line Item Insurance (USD)"].fillna(data["Line Item Insurance (USD)"].median)
```

```python
#Again check for missing values
```

```python
data.isnull().sum()
```

```python
# Array of all the countries
countryArray = data.Country.unique()
print(countryArray,"\n There are total",len(countryArray) ,"unique countries in dataset")
```

```python
#plot count of each country
itemCount = data.Country.value_counts()
itemCount
```

```python
#top 15 countries 
country = itemCount[:15]
country
```

```python
fig = plt.figure(figsize=(17,5))
plt.bar(country.index,country)
plt.ylabel("Count")
plt.xlabel("Countries");
```

```python
shipment_mode = data["Shipment Mode"]
shipment_mode.value_counts()
label = shipment_mode.value_counts().index
```

```python
plt.pie(shipment_mode.value_counts(), labels=label, autopct='%2.1f%%');
plt.title("Shipment Mode");
```

```python
UniqueItem = data['Manufacturing Site'].unique()
print("All Unique Manufacturing Site \n")
print(UniqueItem)
```

```python
plt.figure(figsize=(22,6))
TopTenManufacturingSite=data.groupby('Manufacturing Site').size().nlargest(10)
print(TopTenManufacturingSite)
TopTenManufacturingSite.plot(kind='bar')
plt.title('Top 10 Manufacturing Site \n')
plt.ylabel('Total Count')
plt.xlabel('Manufacturing Site Name');
```
